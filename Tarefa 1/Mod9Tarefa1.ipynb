```python
# Adicionar coluna com o mês
sinasc_raw['MES'] = sinasc_raw['DATA NASCIMENTO'].dt.month

# Agrupar por mês e cidade, contando os nascimentos
nascimentos_por_cidade = sinasc_raw.groupby(['MES', 'munResNome']).size().reset_index(name='TOTAL_NASCIMENTOS')

# Para cada mês, encontrar a cidade com o maior número de nascimentos
cidade_mais_nascimentos = nascimentos_por_cidade.loc[nascimentos_por_cidade.groupby('MES')['TOTAL_NASCIMENTOS'].idxmax()]

cidade_mais_nascimentos
```

```python
# Encontrar a segunda cidade com mais nascimentos por mês
segunda_cidade_mais_nascimentos = nascimentos_por_cidade.groupby('MES').nth(1).reset_index()
segunda_cidade_mais_nascimentos
```

```python
# Encontrar a terceira cidade com mais nascimentos por mês
terceira_cidade_mais_nascimentos = nascimentos_por_cidade.groupby('MES').nth(2).reset_index()
terceira_cidade_mais_nascimentos
```

```python
# Encontrar a quarta cidade com mais nascimentos por mês
quarta_cidade_mais_nascimentos = nascimentos_por_cidade.groupby('MES').nth(3).reset_index()
quarta_cidade_mais_nascimentos
```

Em todo o ano de 2019, durante todos os meses, a cidade com mais nascimento foi Porto Velho, em seguida vem Ji-Paraná, que é a segunda em todos os meses, com exceção de Outubro, onde é a terceira. A terceira posição por mês possui 3 cidades diferentes, e o mesmo acontece para a quarta posição.
